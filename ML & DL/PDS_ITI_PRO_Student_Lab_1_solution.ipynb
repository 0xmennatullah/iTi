# Problem 1

Use numpy to simulate the vector outcomes for repeating the following random experiments $1000$ times and use the relative frequency to compute the probability of the corresponding events:
    
1. **Experiment:** Flipping a fair coin, **Event:** A head occurs.
     
Observe how changing the number of repetition affects the computed probabilities.

**Hints:** 
- Use [numpy.random.randint](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html) or [numpy.random.choice](https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html) to simulate the experiment.
- Other useful functions [numpy.arange](https://numpy.org/doc/stable/reference/generated/numpy.arange.html), [numpy.cumsum](https://numpy.org/doc/stable/reference/generated/numpy.cumsum.html), [numpy.sum](https://numpy.org/doc/stable/reference/generated/numpy.sum.html), and [matplotlib.pyplot.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
#type your code here:
import numpy as np
import matplotlib.pyplot as plt

#encoding head n tail => 0,1
n_trials = 1000
coin_flip = np.random.choice([0,1], size=n_trials, p=[0.5,0.5])

#freq of head
prob_head = np.sum(coin_flip)/n_trials
print(f"prob of head: {prob_head:.4f}")

#increasing filps
trial_range = np.arange(1, n_trials+1)
cuml_prob = np.cumsum(coin_flip)/trial_range

plt.plot(trial_range, cuml_prob)
plt.legend()
plt.show()
```

2. **Experiment:** Rolling a fair die, **Event:** A number greater than 3 occurs.

```python
#type your code here:
n_trials=1000

die_roll = np.random.randint(1,7, size=n_trials)

#n>3
event = (die_roll>3).astype(int)
prob = np.sum(event)/n_trials
print(f"prob of number > 3: {prob:.4f}")

#trial increase
range = np.arange(1, n_trials+1)
cuml_prob = np.cumsum(event)/range

plt.plot(trial_range, cuml_prob)
plt.legend()
plt.show()
```

3. **Experiment:** Rolling two fair dice, **Event:** Sum of two dice equal 5.

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of rolls
n = 1000

# Simulate rolling two dice n times
die1 = np.random.randint(1, 7, size=n)
die2 = np.random.randint(1, 7, size=n)

# Calculate the sum of each pair
sums = die1 + die2

# Mark 1 where sum == 5, else 0
sum_is_5 = np.where(sums == 5, 1, 0)

# Cumulative count of successful outcomes
cumulative_successes = np.cumsum(sum_is_5)

# Trial numbers
trials = np.arange(1, n + 1)

# Estimated probability at each trial
estimated_prob = cumulative_successes / trials

# Plotting the convergence
plt.figure(figsize=(10, 5))
plt.plot(trials, estimated_prob)
plt.axhline(y=4/36, color='red', linestyle='--', label="True P(Sum = 5) = 4/36 ≈ 0.111")
plt.xlabel("Number of Trials")
plt.ylabel("Estimated Probability (Sum = 5)")
plt.title("Convergence of Estimated Probability: Sum of Two Dice = 5")
plt.legend()
plt.grid(True)
plt.show()
```
