```python
import sklearn
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm
import pandas as pd
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
!pip install plotnine
from plotnine import *

from sklearn.cluster import KMeans
```

```python
n_samples = 1500
dataset1 = sklearn.datasets.make_circles(n_samples=n_samples, factor=.5, noise=.05)
dataset1=pd.DataFrame(dataset1[0])

dataset2 = sklearn.datasets.make_moons(n_samples=n_samples, noise=.05)
dataset2=pd.DataFrame(dataset2[0])

dataset3 = sklearn.datasets.make_blobs(n_samples=n_samples, random_state=8)
dataset3=pd.DataFrame(dataset3[0])

z = StandardScaler()
dataset1[["x","y"]] = z.fit_transform(dataset1)
ggplot(dataset1, aes(x = "x", y = "y")) + geom_point() 
```

```python
dataset2[["x","y"]] = z.fit_transform(dataset2)
ggplot(dataset2, aes(x = "x", y = "y")) + geom_point() 
```

```python
dataset3[["x","y"]] = z.fit_transform(dataset3)
ggplot(dataset3, aes(x = "x", y = "y")) + geom_point() 
```

```python
#all data set in 
data = [dataset1 , dataset2 , dataset3]
for i in data :
    plt.scatter( i.iloc[:, 0], i. iloc[:, 1])
    plt.show()
```

```python
for i in data : 
    i[["x", "y"]] = z.fit_transform(i[["x", "y"]])
```

```python
def kmeans_plot(data_set, n_clusters, rand_state):
    km = KMeans(n_clusters=n_clusters, random_state=rand_state)
    km.fit(data_set[["x", "y"]])
    
    pred = km.predict(data_set[["x", "y"]])
    plt.scatter(data_set["x"], data_set["y"], c=pred)
    plt.title("KMeans Clustering")
    plt.show()
```

```python
kmeans_plot(dataset1 , 2 , 0 )
```

```python
def data_plot(dataset):
    plt.scatter(dataset["x"] , dataset["y"])
    plt.title("original")
```

```python
def cluster(data_set , no_cluster , rand_state):
    plt.figure(figsize=(50,15))
    plt.subplot(1,2,1)
    data_plot(data_set)
    plt.subplot(1,2,2)
    kmeans_plot(data_set , no_cluster , rand_state )
    plt.show()
    
    
```

```python
cluster(dataset1 , 2 , 0)
```

```python
cluster(dataset2 , 2 , 0)
```

```python
cluster(dataset3 , 3 , 0)
```
